#### 1.1. Скачать данные

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data.csv', sep=',')
df
```

#### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(7, 7))
plt.hist(df['price'])
plt.title('Распределение недвижимости по стоимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество');
```

#### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(7, 7))
sns.histplot(df['sqft_living'], bins=20)
plt.title('Распределение жилой квадратуры')
plt.xlabel('Жилые кваадратные метры')
plt.ylabel('Количество');
```

#### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(7, 7))
sns.histplot(df['yr_built'], bins=20)
plt.title('Распределение зданий по году постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество');
```

### Условие 2: 2 задача

#### 2.1 Изучите распределение домов от наличия вида на набережную

Постройте график

```python
data = df['waterfront'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(data, autopct='%1.1f%%')
plt.legend(['нет', 'да']);
```

#### 2.2 Изучите распределение этажей домов

```python
data = df['floors'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(data, autopct='%1.1f%%', labels=data.index)
plt.title('Количество этажей');
```

#### 2.3 Изучите распределение состояния домов

```python
data = df['condition'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(data, autopct='%1.1f%%', labels=data.index);
```

### Условие 3: 3 задача

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
sns.jointplot(x=df['price'], y=df['sqft_basement'], kind='reg');
```

Дома с большей общей площадью стоят дороже

```python
plt.figure(figsize=(10, 10))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['sqft_living15'], y=df['grade'], color ='green', kind='reg');
```

Дома с большей жилой площадью имеет более высокие оценки

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5, showfliers=False);
plt.xlabel('Цена')
plt.ylabel('Оценка')
plt.title('Распределение цены в зависимости от оценки дома');
```

Более дорогие дома имеют более высокие оценки

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1, showfliers=False);
plt.xlabel('Цена')
plt.ylabel('Количество этажей')
plt.title('Зависимость цены дома от количества этажей');
```

Наибольшая стоимость у домов с 2,5 этажами

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1, showfliers=False);
plt.xlabel('Цена')
plt.ylabel('Вид')
plt.title('Зависимость цены дома от вида');
```
